# Emerging Technologies Project
### Project Brief/Objective
The Objective of this project is create a web service that uses machine learning to make pre-dictions based on a pre-made dateset. The goal is to produce a model that accurately predicts wind turbine power output from wind speed values, as in the data set



Wind Energy is the most developed technology worldwide with over 597GW capacity in 2018.[1] Wind Turbines are thus, very important in the upkeep of environmentally friendly energy generation.
Despite this importance, wind turbines are notoriously inefficient and unreliable. Wind turbines are unreliable because of the obvious fact that they rely on wind, a naturally occuring condition. If there is no wind, or low-speed wind, then there is next to no power output, and the buildings, farms, houses that rely on the energy provided by the turbines will either have to go about their day with no power, or use other sources of power e.g. fossil fuels, spare generators.
"Wind farms, however, cannot currently provide this reliability. In fact, on the scale of most countries aggregate wind farm output can be assumed to have almost zero reliability. In this sense, every wind farm must have a fossil fuel power plant sitting in wait for when the wind does not blow."[2]


### The Data Set
The data set provided for this project has odd values in it e.g 0 wind speed = 0 power output but also 25 output also equals 0 power output. However inbetween these values, there are many wind speed values that result in a positive power output.
This may be due to the fact that wind turbines will shut off after a certain speed of wind to prevent an energy surge overwhelming the power grid, and also to prevent the blades of the turbine from being damaged, the blades the blades begin to feather, or point into the wind to reduce their surface area. In some instances, although not common, the blades can even be locked down to ride out severe gusts.[3] This means that at high winds, the turbines will not be producing any power output

```python
# Neural networks.
import tensorflow.keras as kr
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.models import Model

# Numerical arrays
import numpy as np

# Data frames.
import pandas as pd

# Plotting
import matplotlib.pyplot as plt

```

```python
# Plot style.
plt.style.use("ggplot")

# Plot size.
plt.rcParams['figure.figsize'] = [14, 8]
```

```python
powerData = pd.read_csv("powerproduction.csv")

powerData
```

For my first model, the model will have the entire raw dataset, with 3 layers;1 sigmoid and 1 linear layer. Sigmoid helps to limit the types of values put through the model. "sigmoid returns a value close to zero, and for large values (>5) the result of the function gets close to 1."[4] 

```python
# Build our model.
model = kr.models.Sequential()
# For a single-input model with 3 classes (binary classification):
model.add(kr.layers.Dense(4, input_shape=(1,), activation='sigmoid', kernel_initializer="ones", bias_initializer="zeros"))

model.add(kr.layers.Dense(3, input_shape=(1,), activation='linear', kernel_initializer="ones", bias_initializer="zeros"))

model.add(kr.layers.Dense(1, activation='linear', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))

model.compile(kr.optimizers.Adam(lr=0.001), loss='mean_squared_error')
```

```python
model.fit(powerData['speed'], powerData['power'],epochs=5000, batch_size=32)
model.predict([1.001])
```

```python
plt.plot(powerData['speed'], powerData['power'], "bo", label='actual')
plt.plot(powerData['speed'], model.predict(powerData['speed']),'r', label='prediction')
plt.legend();
```

```python
model.save("powerprod.h5")
```

This model isn't as accurate, as you can see, the red "prediction" line dips away from the center of the blue "actual" data near the end of the graph, resulting a high loss of around 180 - 220, with an average loss of 200.

If we wanted to create a more accurate model, we could remove all of the outlier data points from the data set i.e. every piece of data with a 0 power output. By doing this, we are removing the irrelvant outlier data and only focusing on wind speeds that actually result in power being output.

```python
powerData.drop(powerData.index[powerData['power']==0], inplace= True)
powerData
```

```python
# Build our model.
modelWithoutZero = kr.models.Sequential()
# For a single-input model with 3 classes (binary classification):
modelWithoutZero.add(kr.layers.Dense(3, input_shape=(1,), activation='sigmoid', kernel_initializer="ones", bias_initializer="zeros"))

modelWithoutZero.add(kr.layers.Dense(3, input_shape=(1,), activation='linear', kernel_initializer="ones", bias_initializer="zeros"))

modelWithoutZero.add(kr.layers.Dense(1, activation='linear', kernel_initializer="glorot_uniform", bias_initializer="glorot_uniform"))

modelWithoutZero.compile(kr.optimizers.Adam(lr=0.001), loss='mean_squared_error')
```

```python
modelWithoutZero.fit(powerData['speed'], powerData['power'],epochs=500, batch_size=32)
modelWithoutZero.predict([1.001])
```

```python
plt.plot(powerData['speed'], powerData['power'], "bo", label='actual')
plt.plot(powerData['speed'], modelWithoutZero.predict(powerData['speed']),'ro', label='prediction')
plt.legend();
```

```python
modelWithoutZero.save("powerprodWithoutZeros.h5")
```

As you can see with the loss being much lower, around 16-17, and with the graph *not* dropping off at the end of the graph, that this model is much more accurate than the first. By removing the outlier data, we have forced the data model to focus on only relevant data, resulting in more accurate predictions on the web service.
